``````python
import sympy as sp

# Declare symbolic variables
a21, a31, a32, a41, a42, a43, b1, b2, b3, b4 = sp.symbols("a21, a31, a32, a41, a42, a43, b1, b2, b3, b4")

# Define c values
c2, c3, c4 = sp.Rational(1,3), sp.Rational(1,2), 1

# Define order conditions
eq1 = b1 + b2 + b3 + b4 - 1
eq2 = b2 * c2 + b3 * c3 + b4 * c4 - sp.Rational(1,2)
eq3 = b2 * c2 ** 2 + b3 * c3 ** 2 + b4 * c4 ** 2 - sp.Rational(1,3)
eq4 = b2 * c2 ** 3 + b3 * c3 ** 3 + b4 * c4 ** 3 - sp.Rational(1,4)
eq5 = b3 * a32 * c2 + b4 * a42 * c2 + b4 * a43 * c3 - sp.Rational(1,6)
eq6 = b3 * a32 * c3 * c2 + b4 * c4 * a42 * c2 + b4 * c4 * a43 * c3 - sp.Rational(1,8)
eq7 = b3 * a32 * c2 ** 2 + b4 * a42 * c2 ** 2 + b4 * a43 * c3 ** 2 - sp.Rational(1,12)
eq8 = c2 - a21
eq9 = c3 - a31 - a32
eq10 = c4 - a41 - a42 - a43

# Solve the order conditions
sp.solve((eq1, eq2, eq3, eq4, eq5, eq6, eq7, eq8, eq9, eq10))
``````

````{exercise}
:label: ch2-ex-rk2-hand-calc

Using pen and paper, apply your Runge-Kutta method derived in {ref}`ch2-ex-rk2-derivation` to solve the following initial value problem using a step length of $h=0.4$

\begin{align*}
    y' = t - y, \qquad t\in [0,2],\qquad y(0)=1
\end{align*}

Write your solutions correct to 4 decimal places.

```{dropdown} Solution

```
````

``````python
import numpy as np
import matplotlib.pyplot as plt

# Define ODE function
def f(t, y):
    return t - y

# Define IVP parameters
tspan = [0, 2]  # boundaries of the t domain
y0 = [1]        # initial values
h = 0.4         # step length

# Print table headers
print("|  t   |    y    |   k1    |   k2    |")
print("|:----:|:-------:|:-------:|:-------:|")
print(f"| {0:0.2f} | {y0[0]:7.4f} |    -    |    -    |")

N = len(y0)
nsteps = int((tspan[1] - tspan[0]) / h)
t = np.zeros(nsteps + 1)
y = np.zeros((N, nsteps + 1))
    
t[0] = tspan[0]
y[:,0] = y0

for n in range(nsteps):
    k1 = f(t[n], y[:,n])
    k2 = f(t[n] + 3/4 * h, y[:,n] + 3/4 * h * k1)
    y[:,n+1] = y[:,n] + h / 3 * (k1 + 2 * k2)
    t[n+1] = t[n] + h
    print(f"| {t[n+1]:0.2f} | {y[0,n+1]:7.4f} | {k1[0]:7.4f} | {k2[0]:7.4f} |")

``````


`````{exercise}
:label: ch2-ex-myrk2

Write a Python of MATLAB program that computes the solution to the IVP from {ref}`ch2-ex-rk2-hand-calc` using the second-order ERK method derived in {ref}`ch2-ex-rk2-derivation`. The exact solution to this IVP is $y = t + 2e^{-t} - 1$, produce a plot of the solution and the exact solution on the same set of axes.

````{dropdown} Solution
```{glue:figure} ex2.8_plot
```
````
`````

``````python
import numpy as np
import matplotlib.pyplot as plt

# Define RK2 method function
def rk2(f, tspan, y0, h):
    
    N = len(y0)
    nsteps = int((tspan[1] - tspan[0]) / h)
    t = np.zeros(nsteps + 1)
    y = np.zeros((N, nsteps + 1))
        
    t[0] = tspan[0]
    y[:,0] = y0

    for n in range(nsteps):
        k1 = f(t[n], y[:,n])
        k2 = f(t[n] + 3/4 * h, y[:,n] + 3/4 * h * k1)
        y[:,n+1] = y[:,n] + h / 3 * (k1 + 2 * k2)
        t[n+1] = t[n] + h

    return t, y.T


# Define exact solution
def exact(t):
    return t + 2 * np.exp(-t) - 1

# Solve the IVP using the Euler method
t, y = rk2(f, tspan, y0, h)

# Calculate exact solution
t_exact = np.linspace(tspan[0], tspan[1], 100)
y_exact = exact(t_exact)

# Plot solution
fig, ax = plt.subplots()
plt.plot(t_exact, y_exact, "k", label="Exact")
plt.plot(t, y[:,0], "b-o", label="RK2")
plt.xlabel("$t$", fontsize=12)
plt.ylabel("$y(t)$", fontsize=12)
plt.title("$y' = t - y$, $y(0) = 1$")
plt.legend()
plt.show()

from myst_nb import glue
glue("ex2.8_plot", fig, display=False)
``````

`````{exercise}
:label: ch2-ex-rk3-rk4-solution

A third-order explicit Runge-Kutta method has the Butcher tableau

$$ \begin{array}{c|ccc}
    0 & 0 \\
    \frac{1}{2} & \frac{1}{2} \\
    1 & -1 & 2 \\
    \hline

Write a Python or MATLAB program which uses this third-order method, and the fourth-order method derived in {ref}`ch2-ex-rk4-derivation`, to solve the IVP from {ref}`ch2-ex-rk2-hand-calc`. Produce a plot the compares the second-, third- and fourth-order solutions to the exact solution.

````{dropdown} Solution
```{glue:figure} ex2.9_plot
```
````
`````

``````python
import numpy as np
import matplotlib.pyplot as plt

def rk3(f, tspan, y0, h):
    
    N = len(y0)
    nsteps = int((tspan[1] - tspan[0]) / h)
    t = np.zeros(nsteps + 1)
    y = np.zeros((N, nsteps + 1))
        
    t[0] = tspan[0]
    y[:,0] = y0

    for n in range(nsteps):
        k1 = f(t[n], y[:,n])
        k2 = f(t[n] + 1/2 * h, y[:,n] + 1/2 * h * k1)
        k3 = f(t[n] + h, y[:,n] + h * (-k1 + 2 * k2))
        y[:,n+1] = y[:,n] + h / 6 * (k1 + 4 * k2 + k3)
        t[n+1] = t[n] + h

    return t, y.T


def rk4(f, tspan, y0, h):
    
    N = len(y0)
    nsteps = int((tspan[1] - tspan[0]) / h)
    t = np.zeros(nsteps + 1)
    y = np.zeros((N, nsteps + 1))
        
    t[0] = tspan[0]
    y[:,0] = y0

    for n in range(nsteps):
        k1 = f(t[n], y[:,n])
        k2 = f(t[n] + 1/3 * h, y[:,n] + 1/3 * h * k1)
        k3 = f(t[n] + 1/2 * h, y[:,n] + h * (1/8 * k1 + 3/8 * k2))
        k4 = f(t[n] + h, y[:,n] + h * (1/2 * k1 - 3/2 * k2 + 2 * k3))
        y[:,n+1] = y[:,n] + h / 6 * (k1 + 4 * k3 + k4)
        t[n+1] = t[n] + h

    return t, y.T


# Define exact solution
def exact(t):
    return t + 2 * np.exp(-t) - 1

# Solve the IVP
t, y_rk2 = rk2(f, tspan, y0, h)
t, y_rk3 = rk3(f, tspan, y0, h)
t, y_rk4 = rk4(f, tspan, y0, h)

# Calculate exact solution
t_exact = np.linspace(tspan[0], tspan[1], 100)
y_exact = exact(t_exact)

# Plot solution
fig, ax = plt.subplots()
plt.plot(t_exact, y_exact, "k", label="Exact")
plt.plot(t, y_rk2[:,0], "b-o", label="RK2")
plt.plot(t, y_rk3[:,0], "r-o", label="RK3")
plt.plot(t, y_rk4[:,0], "g-o", label="RK4")
plt.xlabel("$t$", fontsize=12)
plt.ylabel("$y(t)$", fontsize=12)
plt.title("$y' = t - y$, $y(0) = 1$")
plt.legend()
plt.show()

glue("ex2.9_plot", fig, display=False)
``````

``````python
hvals = [0.4, 0.2, 0.1, 0.05, 0.025]
E_rk2, E_rk3, E_rk4 = [], [], []
for h in hvals:
    t, y_rk2 = rk2(f, tspan, y0, h)
    t, y_rk3 = rk3(f, tspan, y0, h)
    t, y_rk4 = rk4(f, tspan, y0, h)
    idx = np.argmin(abs(t - 2))
    E_rk2.append(abs(exact(2) - y_rk2[idx,0]))
    E_rk3.append(abs(exact(2) - y_rk3[idx,0]))
    E_rk4.append(abs(exact(2) - y_rk4[idx,0]))

# Print table
print("|   h   |   RK2    |   RK3    |   RK4    |")
print("|:-----:|:--------:|:--------:|:--------:|")
for i in range(len(hvals)):
    print(f"| {hvals[i]:0.3f} | {E_rk2[i]:1.2e} | {E_rk3[i]:1.2e} | {E_rk4[i]:1.2e} | ")

# Plot errors
fig, ax = plt.subplots()
plt.loglog(hvals, E_rk2, "b-o", label="RK2")
plt.loglog(hvals, E_rk3, "r-o", label="RK3")
plt.loglog(hvals, E_rk4, "g-o", label="RK4")
plt.xlabel("$h$", fontsize=12)
plt.ylabel("global truncation error", fontsize=12)
plt.title("$y' = t - y$, $y(0) = 1$")
plt.legend()
plt.show()

glue("ex2.10_plot", fig, display=False)

# Estimate order
print(f"RK2: n = {(np.log(E_rk2[0]) - np.log(E_rk2[-1])) / (np.log(hvals[0]) - np.log(hvals[-1])):0.2f}")
print(f"RK3: n = {(np.log(E_rk3[0]) - np.log(E_rk3[-1])) / (np.log(hvals[0]) - np.log(hvals[-1])):0.2f}")
print(f"RK4: n = {(np.log(E_rk4[0]) - np.log(E_rk4[-1])) / (np.log(hvals[0]) - np.log(hvals[-1])):0.2f}")
``````


`````{exercise}
:label: ch2-ex-rk23

Combining Heun's method and Kutta's third-order method gives the following Butcher tableau for an embedded Runge-Kutta method

$$ \begin{array}{c|ccc}
    0 & \\
    \frac{1}{2} & \frac{1}{2} \\
    1 & -1 & 2 \\ \hline
    & \frac{1}{6} & \frac{2}{3} & \frac{1}{6} \\
    & 0 & 1 & 0 
\end{array}$$

where the first row of the $b$ coefficients gives the third-order accurate solution and the second row gives the second-order accurate solution. Write a Python or MATLAB program that solves the initial value problem from {ref}`ch2-ex-rk2-hand-calc` using an absolute tolerance of $10^{-6}$ and a relative tolerance of $10^{-3}$. 

(a) &emsp; Produce a plot that compares the numerical solution with the exact solution.

````{dropdown} Solution

````

(b) &emsp; Determine the number of successful steps, failed steps and the number of times the ODE function was evaluated.

```{dropdown} Solution
There were 24 successful steps, 2 failed steps and 78 function evaluations.
```
`````

``````python
def rk23(f, tspan, y0, atol=1e-6, rtol=1e-3):

    N = len(y0)
    t = np.zeros(100000)
    y = np.zeros((N, 100000)) 
    t[0] = tspan[0]
    y[:,0] = y0
    
    h = 0.8 * rtol ** (1 / 3)
    n, nfail = 0, 0
    while t[n] < tspan[-1]:

        k1 = f(t[n], y[:,n])
        k2 = f(t[n] + 1/2 * h, y[:,n] + 1/2 * h * k1)
        k3 = f(t[n] + h, y[:,n] + h * (-k1 + k2))
        y3 = y[:,n] + h / 6 * (k1 + 4 * k2 + k3)
        y2 = y[:,n] + h * k2

        delta = np.linalg.norm(y3 - y2)
        tol = atol + rtol * np.linalg.norm(y3)
        if delta < tol:
            y[:,n+1] = y3
            t[n+1] = t[n] + h
            n += 1
        else:
            nfail += 1
        
        r = max(0.1, 0.8 * (tol / delta) ** (1/3))
        h = min(r * h, tspan[-1] - t[n])

    print(f"number of successful steps:     {n}")
    print(f"number of failed steps:         {nfail}")
    print(f"number of function evaluations: {(n + nfail) * 3}")

    return t[:n+1], y[:,:n+1].T


# Solve IVP
t, y = rk23(f, tspan, y0, 1e-6, 1e-3)

# Plot solution
fig, ax = plt.subplots()
plt.plot(t_exact, y_exact, "k", label="Exact")
plt.plot(t, y[:,0], "b-o", label="RK3(2)")
plt.xlabel("$t$", fontsize=12)
plt.ylabel("$y(t)$", fontsize=12)
plt.title("$y' = t - y$, $y(0) = 1$")
plt.legend()
plt.show()

glue("ex2.11_plot", fig, display=False)
``````
