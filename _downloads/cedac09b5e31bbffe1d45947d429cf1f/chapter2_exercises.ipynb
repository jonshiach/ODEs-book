## Exercise 2.8

```python
import numpy as np
import matplotlib.pyplot as plt

# Define RK2 method
def rk2(f, tspan, y0, h):
  
    nsteps = int((tspan[1] - tspan[0]) / h)
    m = len(y0)
    t = np.zeros(nsteps + 1)
    y = np.zeros((nsteps + 1, m))
    t[0] = tspan[0]
    y[0,:] = y0

    for n in range(nsteps):
        k1 = f(t[n], y[n,:])
        k2 = f(t[n] + 3/4 * h, y[n,:] + 3/4 * h * k1)
        y[n+1,:] = y[n,:] + h / 3 * (k1 + 2 * k2);
        t[n+1] = t[n] + h

    return t, y


# Define ODE function
def f(t, y):
    return t - y


# Define IVP parameters
tspan = [0, 2]  # boundaries of the t domain
y0 = [1]        # initial values
h = 0.4         # step length

# Solve IVP
t, y = rk2(f, tspan, y0, h)

# Calculate exact solution
def exact(t):
    return t + 2 * np.exp(-t) - 1

texact = np.linspace(tspan[0], tspan[1], 100)
yexact = exact(texact)

# Plot solution
fig, ax = plt.subplots()
plt.plot(texact, yexact, 'k', label='Exact')
plt.plot(t, y, 'b-o', label="RK2")
plt.xlabel('$t$')
plt.ylabel('$y(t)$')
plt.title('$y = t - y$, $y(0) = 1$')
plt.legend()
plt.show()
```

## Exercise 2.9

```python
def rk3(f, tspan, y0, h):
    
    m = len(y0)
    nsteps = int((tspan[1] - tspan[0]) / h)
    t = np.zeros(nsteps + 1)
    y = np.zeros((nsteps + 1, m))
        
    t[0] = tspan[0]
    y[0,:] = y0

    for n in range(nsteps):
        k1 = f(t[n], y[n,:])
        k2 = f(t[n] + 1/2 * h, y[n,:] + 1/2 * h * k1)
        k3 = f(t[n] + h, y[n,:] + h * (-k1 + 2 * k2))
        y[n+1,:] = y[n,:] + h / 6 * (k1 + 4 * k2 + k3)
        t[n+1] = t[n] + h

    return t, y


def rk4(f, tspan, y0, h):
    
    m = len(y0)
    nsteps = int((tspan[1] - tspan[0]) / h)
    t = np.zeros(nsteps + 1)
    y = np.zeros((nsteps + 1, m))
        
    t[0] = tspan[0]
    y[0,:] = y0

    for n in range(nsteps):
        k1 = f(t[n], y[n,:])
        k2 = f(t[n] + 1/3 * h, y[n,:] + 1/3 * h * k1)
        k3 = f(t[n] + 1/2 * h, y[n,:] + h * (1/8 * k1 + 3/8 * k2))
        k4 = f(t[n] + h, y[n,:] + h * (1/2 * k1 - 3/2 * k2 + 2 * k3))
        y[n+1,:] = y[n,:] + h / 6 * (k1 + 4 * k3 + k4)
        t[n+1] = t[n] + h

    return t, y


# Solve the IVP
t, y_rk2 = rk2(f, tspan, y0, h)
t, y_rk3 = rk3(f, tspan, y0, h)
t, y_rk4 = rk4(f, tspan, y0, h)

# Calculate exact solution
def exact(t):
    return t + 2 * np.exp(-t) - 1

t_exact = np.linspace(tspan[0], tspan[1], 100)
y_exact = exact(t_exact)

# Plot solution
fig, ax = plt.subplots()
plt.plot(t_exact, y_exact, "k", label="Exact")
plt.plot(t, y_rk2[:,0], "b-o", label="RK2")
plt.plot(t, y_rk3[:,0], "r-o", label="RK3")
plt.plot(t, y_rk4[:,0], "g-o", label="RK4")
plt.xlabel("$t$", fontsize=12)
plt.ylabel("$y(t)$", fontsize=12)
plt.title("$y' = t - y$, $y(0) = 1$")
plt.legend()
plt.show()
```

## Exercise 2.10

```python
# Loop though h values and calculate global truncation errors
hvals = [0.4, 0.2, 0.1, 0.05]
error_rk2, error_rk3, error_rk4 = [], [], []
for h in hvals:
    t, y_rk2 = rk2(f, tspan, y0, h)
    t, y_rk3 = rk3(f, tspan, y0, h)
    t, y_rk4 = rk4(f, tspan, y0, h)
    idx = np.argmin(abs(t - 2))
    error_rk2.append(abs(exact(2) - y_rk2[idx,0]))
    error_rk3.append(abs(exact(2) - y_rk3[idx,0]))
    error_rk4.append(abs(exact(2) - y_rk4[idx,0]))

# Print table
print("|   h   |   RK2    |   RK3    |   RK4    |")
print("|:-----:|:--------:|:--------:|:--------:|")
for i in range(len(hvals)):
    print(f"| {hvals[i]:0.3f} | {error_rk2[i]:1.2e} | {error_rk3[i]:1.2e} | {error_rk4[i]:1.2e} | ")

# Plot errors
fig, ax = plt.subplots()
plt.loglog(hvals, error_rk2, "b-o", label="RK2")
plt.loglog(hvals, error_rk3, "r-o", label="RK3")
plt.loglog(hvals, error_rk4, "g-o", label="RK4")
plt.xlabel("$h$", fontsize=12)
plt.ylabel("global truncation error", fontsize=12)
plt.legend()
plt.show()

# Estimate order
print(f"RK2: n = {(np.log(error_rk2[0]) - np.log(error_rk2[-1])) / (np.log(hvals[0]) - np.log(hvals[-1])):0.2f}")
print(f"RK3: n = {(np.log(error_rk3[0]) - np.log(error_rk3[-1])) / (np.log(hvals[0]) - np.log(hvals[-1])):0.2f}")
print(f"RK4: n = {(np.log(error_rk4[0]) - np.log(error_rk4[-1])) / (np.log(hvals[0]) - np.log(hvals[-1])):0.2f}")
```

```python
def rk23(f, tspan, y0, atol=1e-6, rtol=1e-3):
  
    m = len(y0)
    t = np.zeros(100000)
    y = np.zeros((100000, m)) 
    t[0] = tspan[0]
    y[0,:] = y0
    
    h = 0.8 * rtol ** (1 / 3)
    n, nfail = 0, 0
    while t[n] < tspan[-1]:

        k1 = f(t[n], y[n,:])
        k2 = f(t[n] + 1/2 * h, y[n,:] + 1/2 * h * k1)
        k3 = f(t[n] + h, y[n,:] + h * (-k1 + k2))
        y3 = y[n,:] + h / 6 * (k1 + 4 * k2 + k3)
        y2 = y[n,:] + h * k2

        delta = np.linalg.norm(y3 - y2)
        tol = atol + rtol * np.min(np.abs(y3))
        if delta < tol:
            y[n+1,:] = y3
            t[n+1] = t[n] + h
            n += 1
        else:
            nfail += 1
        
        r = max(0.1, 0.8 * (tol / delta) ** (1/3))
        h = min(r * h, tspan[-1] - t[n])

    print(f"number of successful steps:     {n}")
    print(f"number of failed steps:         {nfail}")
    print(f"number of function evaluations: {(n + nfail) * 3}")

    return t[:n+1], y[:n+1,:]


# Solve IVP
atol, rtol = 1e-6, 1e-3
t, y = rk23(f, tspan, y0, atol, rtol)

# Plot solution
fig, ax = plt.subplots()
plt.plot(t_exact, y_exact, "k", label="Exact")
plt.plot(t, y[:,0], "b-o", label="RK3(2)")
plt.xlabel("$t$", fontsize=12)
plt.ylabel("$y(t)$", fontsize=12)
plt.title("$y' = t - y$, $y(0) = 1$")
plt.legend()
plt.show()
```
