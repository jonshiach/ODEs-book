````python
from functions import *
import sympy as sp
import subprocess

def lu_latex(A):
    n = A.shape[0]
    L, U = sp.eye(n), sp.zeros(n, n)

    string = ""
    string += "$$ \\begin{align*} \n"

    for j in range(n):
        string += f"    j &= {j + 1}: & "
        for i in range(n):
            sum_ = 0
            if i <= j:
                if (i == 0):
                    string += f"    u_{{{i+1}{j+1}}} &= a_{{{i+1}{j+1}}} "
                else:
                    string += f"               && u_{{{i+1}{j+1}}} &= a_{{{i+1}{j+1}}} "

                for k in range(i):
                    sum_ += L[i,k] * U[k,j]
                    string += f"- \ell_{{{i+1}{k+1}}} u_{{{k+1}{j+1}}} "
        
                U[i,j] = A[i,j] - sum_   

                string += f"= {sp.latex(A[i,j])} "
                for k in range(i):
                    if L[i,k] >= 0:
                        string += f"- {sp.latex(L[i,k])} \\left( {sp.latex(U[k,j])} \\right) "
                    else:
                        string += f" - \\left({sp.latex(L[i,k])} \\right) \\left( {sp.latex(U[k,j])} \\right) "

                if i > 0:
                    string += f"= {sp.latex(U[j,j])}, \\\ \n"
                else:
                    string += ", \\\ \n"
            
            else:  
                string += f"            && \ell_{{{i+1}{j+1}}} &= \\frac{{1}}{{u_{{{j+1}{j+1}}}}} \\left( a_{{{i+1}{j+1}}} "  
                for k in range(j):
                    sum_ += L[i,k] * U[k,j]
                    string += f"- \ell_{{{i+1}{k+1}}} u_{{{k+1}{j+1}}} "
                
                string += f"\\right) = \\frac{{1}}{{{U[j,j]}}} \\left( {sp.latex(A[i,j])} "

                for k in range(j):
                    if L[i,k] >= 0:
                        string += f"- {sp.latex(L[i,k])} \\left( {sp.latex(U[k,j])} \\right) "
                    else:
                        string += f"- \\left( {sp.latex(L[i,k])} \\right) \\left( {sp.latex(U[k,j])} \\right) "
                    
                L[i,j] = (A[i,j] - sum_) / U[j,j]

                string += f"\\right) = {sp.latex(L[i,j])}, \\\ \n"
    
    string += "\\end{align*} $$ \n\n"
    string += f"So $L = {sp.latex(L, mat_delim='(', mat_str='matrix')}$ and $U = {sp.latex(U, mat_delim='(', mat_str='matrix')}$. \n\n"
    
    return L, U, string

    
def forward_sub_latex(L, b):
    n = L.shape[0]
    x = sp.zeros(n, 1)
    x[0] = b[0]
    string = "$$ \\begin{align*} \n"
    string += f"    {sp.latex(L, mat_delim='(', mat_str='matrix')} "
    string += "    \\begin{pmatrix} "
    for i in range(n - 1):
        string += f"y_{{{i+1}}} \\\ "

    string += f"y_{{{n}}} \\end{{pmatrix}} &="
    string += f"{sp.latex(b, mat_delim='(', mat_str='matrix')} \\\ \n"
    
    for i in range(n):
        string += f"    y_{{{i+1}}} &= \\frac{{1}}{{{sp.latex(L[i,i])}}} \\left( {sp.latex(b[i])} "

        sum_ = 0
        for j in range(i):
            sum_ += L[i,j] * x[j]
            
        x[i] = (b[i] - sum_) / L[i,i]

        for j in range(i):
            if L[i,j] >= 0:
                string += f"- {sp.latex(L[i,j])} \\left( {sp.latex(x[j])} \\right) "
            else:
                string += f"+ {sp.latex(-L[i,j])} \\left( {sp.latex(x[j])} \\right) "

        string += f"\\right) = {sp.latex(x[i])} "
        if i < n - 1:
            string += ", \\\ \n"
        else:
            string += ".\n"

    string+= "\\end{align*} $$ \n\n"
    
    return x, string


def back_sub_latex(U, b):
    n = U.shape[0]
    x = sp.zeros(1, n)
    string = "$$ \\begin{align*} \n"
    string += f"    {sp.latex(U, mat_delim='(', mat_str='matrix')} "
    string += "    \\begin{pmatrix} "

    for i in range(n - 1):
        string += f"x_{{{i+1}}} \\\ "
    
    string += f"x_{{{n}}} \\end{{pmatrix}} &="
    string += f"{sp.latex(b, mat_delim='(', mat_str='matrix')} \\\ \n"

    x[-1] = b[-1] / U[-1,-1]
    for i in range(n - 1, -1, -1):
        sum_ = 0
        string += f"    x_{{{i+1}}} &= \\frac{{1}}{{{sp.latex(U[i,i])}}} \\left( {sp.latex(b[i])} "
        for j in range(i + 1, n):
            sum_ += U[i,j] * x[j]
            if U[i,j] >= 0:
                string += f"- {sp.latex(U[i,j])} \\left( {sp.latex(x[j])} \\right) "
            else:
                string += f"+ {sp.latex(-U[i,j])} \\left( {sp.latex(x[j])} \\right) "
            
        x[i] = (b[i] - sum_) / U[i,i]

        string += f"\\right) = {sp.latex(x[i])} "
        if i > 0:
            string += ", \\\ \n"
        else:
            string += ".\n"
        
    string+= "\\end{align*} $$ \n\n"
    
    return x, string

# (a) 
labels = ['a']
A = [sp.Matrix([[2, 3, -1], [4, 9, -1], [0, 3, 2]])]
b = [sp.Matrix([4, 18, 11])]

# (b)
labels.append('b')
A.append(sp.Matrix([[3, 9, 5], [1, 2, 2], [2, 4, 5]]))
b.append(sp.Matrix([20, 3, 4]))

# (c)
labels.append('c')
A.append(sp.Matrix([[1, 0, 3, 2], [3, -2, 5, 1], [4, -1, -2, -3], [0, 2, 0, 3]]))
b.append(sp.Matrix([21, 28, -12, 13]))

# (d)
labels.append('d')
A.append(sp.Matrix([[1, 5, 2, 2], [-2, -4, 2, 0], [3, 1, -2, -1], [-3, -3, 4, -1]]))
b.append(sp.Matrix([-10, 10, -2, 4]))

# Calculate solutions
string = ""
for i in range(len(A)):

    string += f"({labels[i]}) Calculate LU decomposition of the coefficient matrix $A = {sp.latex(A[i], mat_delim='(', mat_str='matrix')}$:\n\n"

    L, U, temp_string = lu_latex(A[i])
    string += temp_string

    string += f"Solving $L \\vec{{y}} = \\vec{{b}}$ using forward substitution \n\n"
    y, temp_string = forward_sub_latex(L, b[i])
    string += temp_string

    string += f"Solving $U \\vec{{x}} = \\vec{{y}}$ using back substitution \n\n"
    x, temp_string = back_sub_latex(U, y)
    string += temp_string

print(string)
subprocess.run("pbcopy", text=True, input=string)
````

````python
def cholesky_latex(A):
    n = A.shape[0]
    L = sp.zeros(n, n)

    string = ""
    string += "$$ \\begin{align*} \n"

    for j in range(n):
        string += f"    j &= {j + 1}: & "
        string += f"\\ell_{{{j+1}{j+1}}} &= \\sqrt{{a_{{{j+1}{j+1}}}"
        sum_ = 0
        for k in range(j):
            sum_ += L[j,k] * L[j,k]
            string += f" - \\ell_{{{j+1}{k+1}}}^2"
        
        L[j,j] = sp.sqrt(A[j,j] - sum_)

        string += "} = \\sqrt{" 
        string += f"{sp.latex(A[j,j])}"

        for k in range(j):
            string += f"- {sp.latex(L[j,k])}^2"

        string += "}"
        string += f" = {sp.latex(L[j,j])}, \\\ \n"

        for i in range(j + 1, n):
            string += f"           && \\ell_{{{i+1}{j+1}}} &= \\frac{{1}}{{ \\ell_{{{j+1}{j+1}}} }} \\left( a_{{{i+1}{j+1}}}"
            sum_ = 0
            for k in range(j):
                sum_ += L[i,k] * L[j,k]
                string += f" - \\ell_{{{i+1}{k+1}}} \\ell_{{{j+1}{k+1}}}"

            L[i,j] = 1 / L[j,j] * (A[i,j] - sum_)

            string += f" \\right) = "
            string += f"\\frac{{1}}{{{sp.latex(L[j,j])}}} \\left( {sp.latex(A[i,j])}"
            for k in range(j):
                string += f" - {sp.latex(L[i,k])} \\left( {sp.latex(L[j,k])} \\right)"

            string += f" \\right) = {sp.latex(L[i,j])}, \\\ \n"

    
    string += "\\end{align*} $$ \n\n"
    string += f"so $L = {sp.latex(L, mat_delim='(', mat_str='matrix')}$ . \n\n"
    
    return L, string


# (a) 
labels = ['a']
A = [sp.Matrix([[16, 16, 4], [16, 25, 10], [4, 10, 6]])]
b = [sp.Matrix([-8, -47, -30])]

# (b)
labels.append('b')
A.append(sp.Matrix([[4, 2, 8], [2, 17, 20], [8, 20, 41]]))
b.append(sp.Matrix([36, 50, 122]))

# (c)
labels.append('c')
A.append(sp.Matrix([[9,-9,0,6], [-9, 25, 8, -10], [0, 8, 8, -2], [-6, -10, -2, 33]]))
b.append(sp.Matrix([12, -116, -58, 91]))

# Calculate solutions
string = ""
for i in range(len(A)):

    string += f"({labels[i]}) Calculate Cholesky decomposition of the coefficient matrix $A = {sp.latex(A[i], mat_delim='(', mat_str='matrix')}$:\n\n"

    L, temp_string = cholesky_latex(A[i])
    string += temp_string

    string += f"Solving $L \\vec{{y}} = \\vec{{b}}$ using forward substitution \n\n"
    y, temp_string = forward_sub_latex(L, b[i])
    string += temp_string

    string += f"Solving $L^\\mathsf{{T}} \\vec{{x}} = \\vec{{y}}$ using back substitution \n\n"
    x, temp_string = back_sub_latex(L.T, y)
    string += temp_string

print(string)
subprocess.run("pbcopy", text=True, input=string)
````

````python
def qrgs_latex(A):
    m, n = A.shape
    Q, R = sp.zeros(m,n), sp.zeros(n)

    string = "$$ \\begin{align*} \n"
    for j in range(n): 

        for i in range(j):
            R[i,j] = Q[:,i].dot(A[:,j])

        sum_ = sp.zeros(m,1)
        for i in range(j):
            sum_ += R[i,j] * Q[:,i]

        u = A[:,j] - sum_

        R[j,j] = u.norm()
        Q[:,j] = u / R[j,j]

        string += f"    j &= {j+1}: & "
        for i in range(j):
            R[i,j] = Q[:,i].dot(A[:,j])
            if i > 0:
                string += "    &&"
            string += f"    r_{{{i+1}{j+1}}} &= \\vec{{q}}_{i+1} \\cdot \\vec{{a}}_{j+1} = {sp.latex(Q[:,i], mat_delim='(', mat_str='matrix')} \\cdot {sp.latex(A[:,j], mat_delim='(', mat_str='matrix')}= {sp.latex(R[i,j])}, \\\ \n"


        if j == 0:
            string += f"\\vec{{u}}_{j+1} &= \\vec{{a}}_{j+1} = {sp.latex(A[:,j], mat_delim='(', mat_str='matrix')}, \\\ \n"
        else:
            string += f"       &&   \\vec{{u}}_{j+1} &= \\vec{{a}}_{j+1}"
            for i in range(j):
                string += f" - r_{{{i+1}{j+1}}} \\vec{{q}}_{i+1}"

            string += f" = {sp.latex(A[:,j], mat_delim='(', mat_str='matrix')}"
            for i in range(j):
                if R[i,j] >= 0:
                    string += f" - {sp.latex(R[i,j])} "
                else:
                    string += f" + {sp.latex(-R[i,j])} "
                    
                string += f" {sp.latex(Q[:,i], mat_delim='(', mat_str='matrix')} "
            
            string += f" = {sp.latex(u, mat_delim='(', mat_str='matrix')}, \\\ \n"

        string += f"            && r_{{{j+1}{j+1}}} &= \\|\\vec{{u}}_{j+1}\\| = {sp.latex(R[j,j])}, \\\ \n"
        string += f"            && \\vec{{q}}_{j+1} &= \\frac{{\\vec{{u}}_{j+1}}}{{r_{{{j+1}{j+1}}}}} = \\frac{{1}}{{{sp.latex(R[j,j])}}} {sp.latex(u, mat_delim='(', mat_str='matrix')} = {sp.latex(Q[:,j], mat_delim='(', mat_str='matrix')}, \\\ \n"
        
    string += "\\end{align*} $$" + "\n\n"
    string += f"so $Q = {sp.latex(Q, mat_delim='(', mat_str='matrix')}$ and $R = {sp.latex(R, mat_delim='(', mat_str='matrix')}$.\n\n"

    return string


# (a) 
labels = ['a']
A = [sp.Matrix([[1, 1], [-1, 0]])]

# # (b)
labels += 'b'
A.append(sp.Matrix([[6, 6, 1], [3, 6, 1], [2, 1, 1]]))

# (c)
labels += 'c'
A.append(sp.Matrix([[1, 2, 1], [1, 4, 3], [1, -4, 6], [1, 2, 1]]))

# Calculate solutions
string = ""

for i in range(len(A)):

    string += f"({labels[i]}) \n\n"

    # Calculate QR decomposition
    string += qrgs_latex(A[i])


print(string)
subprocess.run("pbcopy", text=True, input=string)
````
